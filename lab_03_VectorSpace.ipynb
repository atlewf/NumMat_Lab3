## Problem 1

In this problem we will look at the space of real polynomials of degree $\leq n$:

$$
\mathbb{P}_n(\mathbb{R}) = \left\{f: \mathbb{R}\to\mathbb{R} \quad\biggr\rvert\quad f(x) = \sum_{k=0}^n a_k x^k \quad \textrm{for some }a_1, ..., a_n \in \mathbb{R}\right\}.
$$
Note that each point in the space is a function, namely a polynomial.
### a)
Write down the definition of a real vector space.

<font color ='blue'>
Solution: 
</font>
A vector space V over a field F is closed under finite vector addition and scalar multiplication. Also the following conditions must hold for all vectors and scalars $\in \mathbb{R}$:

- Commutativity: $X + Y = Y + X$
- Associativity of vector addition: $(X + Y) + Z = X + (Y + Z)$
- Additive identity: For all vectors in V: $0 + X = X + 0 = X$
- Existence of additive inverse: $X + (-X) = 0$
- Associativity of scalar multiplication: $p(qX) = (pq)X$
- Distributivity of scalar sums $(n + p)Z = nZ + pZ$
- Distributivity of vector sums $s(X + Y) = sX + sY$
- Scalar multiplication identity $1X = X$



### b)

Show that $\mathbb{P}_n(\mathbb{R})$ is a real vector space.

*Hint: Recall the how we define adding two function, and how we define multiplying with a scalar.*




<font color ='blue'>
Solution:
</font>
Let f and g be two functions that are part of the vector space V. Then we check against the eight requirements:


- 1-4: $f(x) + g(x) = f(x) = \sum_{k=0}^n a_k x^k \quad + \sum_{k=0}^m b_k x^k \quad\textrm{for some }a_1,b_1,..., a_n,b_m \in \mathbb{R}$

Easy to see that we can simply change places, add in another function q, add 0 or that f + (-f) = 0. This means that the first four requirements are fulfilled.

- 5: $p \cdot(f(x) \cdot q) = pq \cdot f(x)$
- 6: $(n+p) \cdot f(x) = nf(x) + pf(x)$
- 7: $s \cdot(f(x) + g(x)) = sf(x) + sg(x)$
- 8: $1 \cdot f(x) = f(x)$
    


### c)

Show that $\mathbb{P}_n(\mathbb{R})$ has dimension $n+1$.

*Hint 1: Find a set of polynomials you think is a basis for $\mathbb{P}_n(\mathbb{R})$. Verify that it is in fact a basis.*



*Hint 2: If $p \in \mathbb{P}_n(\mathbb{R})$ and $p \neq 0$, how many roots can $p$ have at most? Recall the fundamental theorem of algebra.*


<font color ='blue'>
    Solution:
</font>    
Let $f \in \mathbb{R}_n$, then the polynomials $f, f', f'',...,f^n$ will form a basis for $\mathbb{P}_n(\mathbb{R})$. This is true if and only if $a_0f + a_1f' + a_2f'' + ... + a_nf^n = 0$. Given that f is a polynomial of degree n, and f' does not have a term of degree n, $a_0$ must be 0. This logic can be repeated for every term $a_0, a_1, a_2, .... , a_n = 0$
    
This means that n+1 elements ${f,f',f'', ... , f^n}$ is a basis for  $\mathbb{P}_n(\mathbb{R})$, and they are all linearly independent. Thus $\mathbb{P}_n(\mathbb{R})$ must have dimension $n+1$

### d)
Write down the definition of a norm on a vector space.

<font color ='blue'>
    Solution:
    
</font>
A norm on a vector space V is a mapping $\Vert \cdot \Vert : V \rightarrow \mathbb{R} $ with the following properties: 

For all $\alpha \in \mathbb{C} $, and all vectors $\textbf{u, v} \in V$:

1) $\Vert \textbf{u}+\textbf{v} \Vert \leq \Vert \textbf{u} \Vert + \Vert \textbf{v} \Vert \$

2) $\Vert \alpha \textbf{u} \Vert = \vert \alpha \vert \Vert \textbf{u} \Vert $

3) $\Vert \textbf{u} \Vert \geq 0$, and $\Vert \textbf{u} \Vert = 0 \iff \textbf{u}=\textbf{0}$ 


### e)
Let $x_0,...,x_n$ be $n+1$ different real numbers. For a polynomial $q \in \mathbb{P}_n(\mathbb{R})$ we now define
$$
\|q\|_P := \sum_{k=0}^n|q(x_k)|.
$$
Show that $\|\cdot\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R}).$



<font color ='blue'>
    Solution:
    
</font>
Showing using the properties from definition. For $u, v \in \mathbb{P}_n(\mathbb{R})$:

1) $\Vert u + v \Vert _P  = \sum_{i=0}^n \vert u(x_n)+v(x_n) \vert \leq \sum_{i=0}^n \left( \vert u(x_n) \vert + \vert v(x_n) \vert \right) =  \sum_{i=0}^n \vert u(x_n) \vert + \sum_{i=0}^n \vert v(x_n) \vert = \Vert u \Vert _P + \Vert v \Vert _P $

2) $\Vert \alpha u \Vert _P  = \sum_{i=0}^n \vert \alpha u(x_n) \vert = \sum_{i=0}^n \vert \alpha \vert \vert u(x_n) \vert = \vert \alpha \vert \sum_{i=0}^n \vert u(x_n) \vert = \vert \alpha \vert \Vert u \Vert _P $

3) Clearly $\Vert u \Vert _P = \sum_{i=0}^n \vert u(x_n) \vert \geq 0$. Further, $\sum_{i=0}^n \vert u(x_n) \vert = 0 \iff \vert u(x_i) \vert = 0$ for all $i=0 \ldots n$ As the polynomial is of degree $n$, having to be equal to zero in $n+1$ points, it must be zero for all possible values of $x_i$. Thus, equality happens only for the zero polynomial $u(x)=0$. 

$\Vert q \Vert _P$ fullfills all three properties as shown, and is therefore a norm on $\mathbb{P}_n(\mathbb{R}).$

### f)
We now define 
$$
\|q\|_{C^n} := \max_{0\leq j \leq n}|q^{(j)}(0)|
$$
where $$q^{(j)} = \dfrac{d^jq}{dx^q}.$$

Show that $\|\cdot\|_{C^n}$ is also a norm on $\mathbb{P}_n(\mathbb{R})$.

<font color ='blue'>
    Solution:
    
</font>
Showing using the properties from definition of norm. For $u, v \in \mathbb{P}_n(\mathbb{R})$:

1) $\Vert u + v \Vert _{C^n}  = \max\limits_{0 \leq j \leq n}|u^{(j)}(0) + v^{(j)}(0)| \leq \max\limits_{0 \leq j \leq n} |u^{(j)}(0)| + \max\limits_{0 \leq j \leq n}|v^{(j)}(0)| = \Vert u \Vert _{C^n} + \Vert v \Vert _{C^n} $

2) $\Vert \alpha u \Vert _{C^n} = \max\limits_{0 \leq j \leq n}| \alpha u^{(j)}(0) | = \max\limits_{0 \leq j \leq n} | \alpha | |u^{(j)}(0)| = |\alpha| \max\limits_{0 \leq j \leq n}| u^{(j)} | = |\alpha| \Vert u \Vert _{C^n} $

3) Clearly $ \|u\|_{C^n} := \max\limits_{0\leq j \leq n}|u^{(j)}(0)| \geq 0$. Further, $\max\limits_{0\leq j \leq n}|u^{(j)}|(0)=0 \iff u^{(j)}(0) = 0$ for all $j=0, \ldots, n$. When having a polynomial of order $n$ with coefficients $c_i$, e.g, $c_n x^n + c_{n-1} x^{n-1} \ldots c_1 x +c_0$, for the n'th derivative to equal zero we need $c_n=0.$ Same goes for all the other coefficients. Thus, equality only occurs for the zero polynomial $u(x)=0$.

$\Vert q \Vert _{C^n}$ fullfills all three properties as shown, and is therefore a norm on $\mathbb{P}_n(\mathbb{R}).$
